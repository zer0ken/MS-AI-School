## üîó ÏÑ†Ìòï ÌöåÍ∑Ä

```python
from sklearn.model_selection import train_test_split

# Î≥ÄÏàòÏÑ†ÌÉù
X = df[['weather', 'temp', 'humidity', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek', 'holiday']]
y = df['count']

# ÌõàÎ†®/ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖã Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, train_size=0.75)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

# Î™®Îç∏ ÏÉùÏÑ±
model = LinearRegression()
model.fit(X_train, y_train)

train_r2_5 = model.score(X_train, y_train)
test_r2_5 = model.score(X_test, y_test)

print(
    f'Train R2: {train_r2_5}\n'
    f'Test R2\t: {test_r2_5}\n'
)

# ÌÖåÏä§Ìä∏Îç∞Ïù¥ÌÑ∞Î°ú ÏòàÏ∏°
pred = model.predict(X_test)

# Î™®Îç∏ ÏÑ±Îä• ÌèâÍ∞Ä
mse = mean_squared_error(y_test, pred)
rmse = root_mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)

print(
    f'MSE\t: {mse:.4f}\n'
    f'RMSE\t: {rmse:.4f}\n'
    f'MAE\t: {mae:.4f}\n'
)
```

### ‚û∞ Ï¢ÖÏÜçÎ≥ÄÏàò Î°úÍ∑∏ Î≥ÄÌôò

```python
from sklearn.model_selection import train_test_split

display(df.head())

# Î≤îÏ£ºÌòï(Categorical) Î≥ÄÏàò Ïõê-Ìï´ Ïù∏ÏΩîÎî©
X = df[['weather', 'temp', 'humidity', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek', 'holiday']]
X_one_hot = pd.get_dummies(X, columns=['year', 'month', 'day', 'hour', 'weather', 'dayofweek'])
display(X_one_hot.head())

y = df['count']

# Î≥ÄÏàò Î°úÍ∑∏Î≥ÄÌôò
y_log = np.log1p(y)
y_log.hist()
plt.show()

# ÌõàÎ†®/ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖã Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(X_one_hot.values, y_log, random_state=42, train_size=0.75)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

# Î™®Îç∏ ÏÉùÏÑ±
model = LinearRegression()
model.fit(X_train, y_train)

train_r2_5 = model.score(X_train, y_train)
test_r2_5 = model.score(X_test, y_test)

print(
    f'Train R2: {train_r2_5}\n'
    f'Test R2\t: {test_r2_5}\n'
)

# ÌÖåÏä§Ìä∏Îç∞Ïù¥ÌÑ∞Î°ú ÏòàÏ∏°
pred = model.predict(X_test)

# Î°úÍ∑∏ Ïó≠Î≥ÄÌôò
pred = np.expm1(pred)
y_test = np.expm1(y_test)

# Î™®Îç∏ ÏÑ±Îä• ÌèâÍ∞Ä
mse = mean_squared_error(y_test, pred)
rmse = root_mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)

print(
    f'MSE\t: {mse:.4f}\n'
    f'RMSE\t: {rmse:.4f}\n'
    f'MAE\t: {mae:.4f}\n'
)
```

```python
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
model = LinearRegression()

default = cross_val_score(model, X, y, scoring='r2', cv=kf).mean()
one_hot = cross_val_score(model, X_one_hot, y, scoring='r2', cv=kf).mean()
one_hot_log = cross_val_score(model, X_one_hot, y_log, scoring='r2', cv=kf).mean()
log = cross_val_score(model, X, y_log, scoring='r2', cv=kf).mean()

print(
    f'Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞\t\t: {default}\n'
    f'Ïõê Ìï´ Ïù∏ÏΩîÎî©\t\t: {one_hot}\n'
    f'Ïõê Ìï´ Ïù∏ÏΩîÎî© Î°úÍ∑∏ Î≥ÄÌôò\t: {one_hot_log}\n'
    f'Î°úÍ∑∏ Î≥ÄÌôò\t\t: {log}\n'
)
```
